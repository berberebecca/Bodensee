```python
#Alternative: Für 2024 scrape ich die Monatsdurchschnittswerte für den Niederschlag
url = "https://www.wetterkontor.de/de/wetter/deutschland/monatswerte-station.asp?id=10929#erlaeuterungen"
from bs4 import BeautifulSoup
html = requests.get(url).text
soup = BeautifulSoup(html)
```

```python
headers = [th.text.strip() for th in soup.select("thead > tr > th")[:4]]
headers
```

```python
first_table = soup.select('table')[0]
```

```python
first_table
```

```python
rows = first_table.find_all('tr')

# Headers neu
headers = []
for th in rows[0].find_all('th'):
    # Clean up the text (remove newlines and extra spaces)
    header = ' '.join(th.text.split())
    headers.append(header)

# Jetzt die Daten ab der zweiten Reihe
data = []
for row in rows[1:]:  
    cols = row.find_all('td')
    if cols:
        row_data = []
        for col in cols:
            # Spaces etc wegmachen
            value = ' '.join(col.text.split())
            row_data.append(value)
        data.append(row_data)


print("Headers:", headers)
print("\nDie ersten paar Zeilen:")
for row in data[:3]:  # Print first 3 rows as example
    print(row)
```
